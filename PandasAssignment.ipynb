```python
import pandas as pd
import numpy as np

df = pd.DataFrame({'From_To': ['LoNDon_paris', 'MAdrid_miLAN','londON_StockhOlm','Budapest_PaRis', 'Brussels_londOn'],
'FlightNumber': [10045, np.nan, 10065, np.nan, 10085],
'RecentDelays': [[23, 47], [], [24, 43, 87], [13], [67, 32]],
'Airline': ['KLM(!)', '<Air France> (12)', '(British Airways. )','12. Air France','"Swiss Air"']})
print(df.dtypes)

#start Q-1

flight_no=df['FlightNumber'][0]
for indx in df.index:
    df['FlightNumber'][indx]=flight_no
    flight_no+=10    

df=df.astype({'FlightNumber': 'int32'})
print('\nAfter conversion\n\n',df.dtypes)
df
#End Q-1

#---------------------------------------------------------------

#Start Q-2
dfTemp=pd.DataFrame(df['From_To'].str.split('_',expand=True))
dfTemp.columns=['From','To']
dfTemp
#End Q-2
#---------------------------------------------------------------
#Start Q-3
dfTemp['From']=dfTemp['From'].str.capitalize()
dfTemp['To']=dfTemp['To'].str.capitalize()
dfTemp
#End Q-3

#---------------------------------------------------------------

#Start Q-4
df.drop('From_To',axis=1,inplace=True)
dfNew=pd.concat([dfTemp,df],axis=1)
dfNew
#End Q-4
#---------------------------------------------------------------

#Start Q-5

dfNew['delay_1']=np.nan
dfNew['delay_2']=np.nan
dfNew['delay_3']=np.nan
ls=dfNew['RecentDelays'].tolist() 


for indx in dfNew.index:
    if(len(ls[indx])==3):
        dfNew['delay_1'][indx]=int(ls[indx][0])
        dfNew['delay_2'][indx]=ls[indx][1]
        dfNew['delay_3'][indx]=ls[indx][2]        
    elif(len(ls[indx])==2):
        dfNew['delay_1'][indx]=ls[indx][0]
        dfNew['delay_2'][indx]=ls[indx][1]
    elif(len(ls[indx])==1):
        dfNew['delay_1'][indx]=ls[indx][0]
         
    #print(ls[indx])
dfNew.drop('RecentDelays',axis=1,inplace=True)
print(dfNew.dtypes)
dfNew
#End Q-5
```
